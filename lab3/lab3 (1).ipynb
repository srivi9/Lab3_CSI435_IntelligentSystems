Below is the simulated annealing algorithm

```python
# simulated annealing algorithm
def simulated_annealing(objective, bounds, n_iterations, step_size, temp):
	#this will give an initial point
	best = bounds[:, 0] + rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])
	#this will evaluate the initial point
	best_eval = objective(best)
	# current working solution
	curr, curr_eval = best, best_eval
	scores = list()
	# this is to run the algorithm
	for i in range(n_iterations):
		# this is to take a step
		candidate = curr + randn(len(bounds)) * step_size
		#this is to evaluate the candidate point
		candidate_eval = objective(candidate)
		#this is to check for the new best solution
		if candidate_eval < best_eval:
			#this will store the new best point
			best, best_eval = candidate, candidate_eval
			#this will keep track of scores
			scores.append(best_eval)
			#this will report progress
			print('>%d f(%s) = %.5f' % (i, best, best_eval))
		#this is to calculate the difference between candidate and current point evaluation
		diff = candidate_eval - curr_eval
		#this will calculate the temperature for current epoch
		t = temp / float(i + 1)
		#this will calculate metropolis acceptance criterion
		metropolis = exp(-diff / t)
		#this will check if we should keep the new point
		if diff < 0 or rand() < metropolis:
			#this will store the new current point
			curr, curr_eval = candidate, candidate_eval
	return [best, best_eval, scores]
```

Below is code to seed the pseudorandom number generator as well as code to initialise bounds,iterations,step size and temperature

```python
#this is to seed the pseudorandom number generator
seed(1)
#this will define an input range
bounds = asarray([[-5.0,5.0]])
#this will define the total iterations
n_iterations = 1000
#this will define the maximum step size
step_size = 0.1
#initial temperature
temp = 10
```

1) How the annealing schedule works?
The annealing schedule controls how the temperature decreases over time.First the temperature is set to a high value so that the algorithm can explore many possible solutions including the worse ones. As the algorithm progresses, the temperature will gradually decrease according to this formula t = temp / float(i + 1) where temp is the initial temperature and i is the current iteration. At lower temperatures, the algorithm becomes more selective so that it can focus on refining the best solutions rather than finding new potential worse ones.

2)How decisions are made regarding whether to use a bad move?
These decisions are made using the metropolis acceptance criterion.This calculates the probability of accepting a worse solution based on the difference in scores and the current temperature.The formula is as follows metropolis = exp(-diff / t). From this formula it is obvious that a higher temperature or a smaller difference is more likely to accept a worse move. The algorithm will then use a random number to decide if the worse solution should be accepted based on this probability.
